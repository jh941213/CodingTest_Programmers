## 1. 자연수 뒤집어 배열로 만들기

image.png

```python
import time 

start = time.time()


def solution(n):
# 답안 작성 부분 ===============
    n = str(n)
    answer=[]
    for i in n[::-1]:
        answer.append(int(i))



# ==============================

    return answer

print(solution(12345))


end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time 

start = time.time()


def solution(absolutes, signs):
# 답안 작성 부분 ===============
    
    answer = 0
    for i in range(len(absolutes)):
        if signs[i]: #true, false 로 트루면 더해준다
            answer += absolutes[i]
        else:
            answer -= absolutes[i]

# ==============================

    return answer

print(solution([4,7,12],[True,False,True] ))
print(solution([1,2,3],[False,False,True]  ))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 없는 숫자 더하기

image.png

```python
import time 

start = time.time()

def solution(number):
# 답안 작성 부분 ===============
    answer = []
    for i in range(10): # 0~9를 비교위한 반복문
        if i not in number:
            answer.append(i)
    
    answer = sum(answer)

# ==============================

    return answer

print(solution([1,2,3,4,6,7,8,0]))
print(solution([5,8,4,0,6,7,9]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time 

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    answer = []
    text_dict={}
    
    for idx, word in enumerate(s):
        if word not in text_dict:
            answer.append(-1)
            text_dict[word] = idx
        else:
            answer.append(idx - text_dict[word])
            text_dict[word] = idx
# ==============================

    return answer

print(solution("banana"))
print(solution("foobar"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 두 개 뽑아서 더하기


image.png

```python
x=[2,3,4,5,6]
len(x)
for i in range(len(x)):
    for j in range(i+1,len(x)):
        print(x[i],x[j])
```

```python
import time 

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    answer=[]
    for i in range(len(s)): #0~4
        for j in range(i+1, len(s)):# 1~4 , 2~4 ,3~4 5~4
            if s[i]+s[j] not in answer: # 더한후 리스트에 없는 값이 면 append
                answer.append(s[i]+s[j])
            
            answer.sort()
# ==============================

    return answer

print(solution([2,1,3,4,1]))
print(solution([5,0,2,7]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
