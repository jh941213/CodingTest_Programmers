```python
max(total)
```

```python
import time 

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    s = s.split(' ')
    total = []
    for i in s:
      total.append(int(i))
    total.sort()                              
    
# ==============================

    return str(total[0]) + " " + str(total[-1])  

print(solution("1 2 3 4"))
print(solution("-1 -2 -3 -4"))
print(solution("-1 -1"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 최솟값 만들기

image.png

```python
import time 

start = time.time()

def solution(A, B):
# 답안 작성 부분 ===============
    answer = 0
    A.sort(reverse = True)  #A 뒤집은다음 곱해서 더해라
    B.sort()
    for i in range(len(A)):
        answer += (A[i]*B[i])
    
# ==============================

    return answer

print(solution([1, 4, 2], [5, 4, 4]))
print(solution([1,2], [3,4]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. JadenCase 문자열 만들기


image.png

```python
#공백을 기준으로 앞글자의 대문자를 해라
```

```python
import time 

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    s = s.split(' ')
    answer =' '
    
    for i in s:
      answer += i.capitalize() + ' '
# ==============================

    return answer

print(solution("3people unFollowed me"))
print(solution("for the last week"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
