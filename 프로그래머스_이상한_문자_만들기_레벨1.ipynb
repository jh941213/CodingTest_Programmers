# Lv.1 이상한 문자 만들기




스크린샷 2023-03-17 오전 9.29.36.png

```python
def solution(s):
    answer = ''
    new_list = s.split(' ')
    for i in new_list:
        for j in range(len(i)):
            if j % 2 == 0:
                answer += i[j].upper()
            else:
                answer += i[j].lower()
        answer+= ' '
    return answer[0:-1]
```

```python
solution("try hello world")
```
