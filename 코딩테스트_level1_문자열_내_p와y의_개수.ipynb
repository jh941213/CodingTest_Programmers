## 프로그래머스 레벨 1 문자열 내 p와 y의 개수

스크린샷 2023-01-27 오전 9.31.05.png

```python
s='zzCCooccCSss'
upper=[]
lower=[]
for i in s:
    if i.isupper():
        upper.append(i)
    else:
        lower.append(i)
        
if len(upper)==len(lower):
    print("False")
else:
    print("True")

```

Solution

```python
def solution(s):
    return s.lower().count('p') == s.lower().count('y')
```
