## 서울에서 김서방 찾기



스크린샷 2023-02-01 오전 9.14.42.png

```python
def solution(seoul):
    
    for i in range(len(seoul)) :
        if seoul[i] == "Kim" :
            answer = i
            
    return('김서방은 ' + str(answer) + '에 있다')
```

```python
solution(3)
```
