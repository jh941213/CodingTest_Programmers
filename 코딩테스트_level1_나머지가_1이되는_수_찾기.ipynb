## 나머지가 1이 되는 수 찾기



스크린샷 2023-01-30 오전 9.23.10.png

```python
def solution(n):
    for i in range(1,n): #1 ~ n 까지 반복문을 돌림
        if n%i == 1:  # n으로 나눈 나머지가 1 인경우 answer = i에 넣어주고 리턴 시킨다
            answer =i
            break     # 반복문을 멈춘다
    return answer
```

```python
solution(5)
```
