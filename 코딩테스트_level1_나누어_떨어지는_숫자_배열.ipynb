## 나누어 떨어지는 숫자 배열



스크린샷 2023-02-01 오전 9.27.32.png

```python
def solution(arr, divisor):
    answer = []
    arr.sort()
    for i in arr:
        if i % divisor ==0:
            answer.append(i)
    return answer if len(answer)!=0 else [-1]
```

```python
solution([5,9,7,10],5)
```
