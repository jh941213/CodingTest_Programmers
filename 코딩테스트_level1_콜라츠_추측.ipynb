## 콜라츠 추측


스크린샷 2023-01-30 오전 9.44.22.png

```python
def solution(num):
    answer = 0
    while num != 1:
        if num % 2 == 0 :
            num //= 2
        else :
            num = (num * 3) + 1
        answer += 1
        if answer >= 500 :
            answer = -1
            break
    return answer
```

```python
solution(3)
```
