## 코딩테스트 자릿수 더하기


스크린샷 2023-01-24 오후 2.27.40.png

```python
def solution(n):
    tot = []
    tot_answer = 0
    for i in str(n):
        tot.append(i)
    
    for i in tot:
        tot_answer += int(i)

    
    return tot_answer
```

```python
solution(456)
```
