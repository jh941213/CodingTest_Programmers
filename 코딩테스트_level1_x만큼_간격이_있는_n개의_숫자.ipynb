## 프로그래머스 레벨 1 x만큼 간격이 있는 N개의 숫자

스크린샷 2023-01-27 오전 9.45.50.png

```python
def solution(x, n):
    answer = []
    
    for i in range(n):
        
        answer.append(x)
        x=x+x

    return answer
```

Solution

```python
def solution(x, n):
    # 함수를 완성하세요
    return [i * x + x for i in range(n)]
```
