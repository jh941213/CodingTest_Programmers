# LV.2 구명보트

```python
def solution(people, limit) :
    answer = 0
    people.sort()

    a = 0 #리스트 1번쨰
    b = len(people) - 1 #리스트 4번쨰
    while a < b : #b가 a보다 큰 경우
        if people[b] + people[a] <= limit : # a와 b를더해서 지정초과보다 작다면
            a += 1 #a는 한칸씩 이동
            answer += 1 #answer 에 1을 더해준다 
        b -= 1 # b는 한칸 줄며 while 문 반복
    return len(people) - answer # 남아있는 둘이 못타는 몸무게들을 리턴 하기 위한 answer 를 뺴준다
```

```python
solution([70, 50, 80, 50], 100)
```
